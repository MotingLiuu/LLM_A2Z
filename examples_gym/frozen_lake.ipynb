```python
import gymnasium as gym
import matplotlib.pyplot as plt
from IPython.display import display, clear_output # notebook
import time

#  'rgb_array' 
env = gym.make('FrozenLake-v1', render_mode='rgb_array')

# 
observation, info = env.reset()

#  RGB 
rgb_array = env.render()

#  Jupyter Notebook 
plt.imshow(rgb_array)
plt.title("FrozenLake Initial State")
plt.axis('off') # 
plt.show() # cell
```

```python
fig, ax = plt.subplots(1, 1, figsize=(4, 4)) # matplotlib
ax.set_axis_off()

for i in range(50): # 50
    action = env.action_space.sample() # 
    observation, reward, terminated, truncated, info = env.step(action)
    rgb_array = env.render()

    ax.imshow(rgb_array)
    ax.set_title(f"FrozenLake - Step {i+1}")
    
    clear_output(wait=True) # 
    display(fig) # 
    time.sleep(0.5) # 

    if terminated or truncated:
        print(f"Episode finished at step {i+1}.")
        break

plt.close(fig) # matplotlib
env.close()
```
