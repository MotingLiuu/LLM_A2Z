```python
import gymnasium as gym
import matplotlib.pyplot as plt
from IPython.display import display, clear_output # 导入用于在notebook中更新输出的工具
import time

# 创建环境，设置为 'rgb_array' 模式
env = gym.make('FrozenLake-v1', render_mode='rgb_array')

# 重置环境
observation, info = env.reset()

# 获取初始帧的 RGB 数组
rgb_array = env.render()

# 在 Jupyter Notebook 中显示图像
plt.imshow(rgb_array)
plt.title("FrozenLake Initial State")
plt.axis('off') # 不显示坐标轴
plt.show() # 这会在当前cell下显示图像
```

```python
fig, ax = plt.subplots(1, 1, figsize=(4, 4)) # 创建一个matplotlib图和轴
ax.set_axis_off()

for i in range(50): # 模拟50步
    action = env.action_space.sample() # 随机动作
    observation, reward, terminated, truncated, info = env.step(action)
    rgb_array = env.render()

    ax.imshow(rgb_array)
    ax.set_title(f"FrozenLake - Step {i+1}")
    
    clear_output(wait=True) # 清除上一个输出
    display(fig) # 显示当前图
    time.sleep(0.5) # 暂停一小会儿，以便观察动画

    if terminated or truncated:
        print(f"Episode finished at step {i+1}.")
        break

plt.close(fig) # 关闭matplotlib图，避免在最后显示空白图
env.close()
```
